```python

# Inicializamos listas para almacenar la información
jornadas = []
fechas = []
partidos = []

# Iteramos sobre los datos
for dato in data:
    # Si el dato tiene el formato de fecha y hora, lo añadimos a la lista de fechas
    if '/' in dato and ':' in dato:
        fechas.append(dato)
    # Si no, si el dato es un número, es el número de jornada
    elif dato.isdigit():
        jornadas.append(dato)
    # Si no es ni fecha ni número, es un partido
    else:
        partidos.append(dato)
```
